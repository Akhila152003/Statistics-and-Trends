# Categorical Graph Function (Bar Chart)

```python
import matplotlib.pyplot as plt
import seaborn as sns


def plot_categorical_graph(df):
    """
    Creates subplots showing the distribution of selected categorical columns in the dataset.

    Parameters:
        df (DataFrame): The DataFrame containing the data.

    Saves:
        'categorical_graphs.png': A set of bar charts for selected columns.
    """
    selected_columns = [
        'MonsoonIntensity', 'TopographyDrainage', 'RiverManagement',
        'Deforestation', 'Urbanization', 'ClimateChange',
        'DamsQuality', 'Siltation', 'AgriculturalPractices'
    ]

    num_cols = len(selected_columns)
    num_rows = (num_cols + 2) // 3  # Arrange 3 columns per row

    fig, axes = plt.subplots(num_rows, 3, figsize=(15, num_rows * 4))
    axes = axes.flatten()  # Flatten the 2D array of axes

    for i, col in enumerate(selected_columns):
        sns.countplot(data=df, x=col, ax=axes[i], palette='viridis', hue=None)  # Set hue to None
        axes[i].set_title(f'Distribution of {col}')
        axes[i].set_xlabel(col)
        axes[i].set_ylabel('Count')
        plt.xticks(rotation=45)  # Rotate x labels for better readability

    # Hide any unused subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.savefig('categorical_graphs.png')
    plt.show()

# Call the function
plot_categorical_graph(df)
```

# Relational Graph Function (Scatter Plot)

```python
def plot_relational_graph(df):
    """
    Creates scatter plots to show the relationship between multiple features and Flood Probability.

    Parameters:
        df (DataFrame): The DataFrame containing the data.

    Saves:
        'relational_graphs.png': A set of scatter plots for selected features vs. Flood Probability.
    """
    selected_features = [
        'Urbanization', 'Deforestation', 'RiverManagement',
        'ClimateChange', 'TopographyDrainage'
    ]

    num_cols = len(selected_features)
    num_rows = (num_cols + 1) // 2  # Arrange 2 columns per row

    fig, axes = plt.subplots(num_rows, 2, figsize=(15, num_rows * 5))
    axes = axes.flatten()  # Flatten the 2D array of axes

    for i, feature in enumerate(selected_features):
        sns.scatterplot(data=df, x=feature, y='FloodProbability', ax=axes[i], alpha=0.6)
        axes[i].set_title(f'{feature} vs. Flood Probability')
        axes[i].set_xlabel(feature)
        axes[i].set_ylabel('Flood Probability')
        axes[i].set_ylim(0, 1)  # Set y limits for better readability

    # Hide any unused subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.savefig('relational_graphs.png')
    plt.show()

# Call the function
plot_relational_graph(df)
```

# Statistical Graph Function (Heatmap)

```python
def plot_statistical_graph(correlation_matrix):
    """
    Creates a heatmap for the correlation matrix of the dataset.

    Parameters:
        correlation_matrix (DataFrame): The correlation matrix of the DataFrame.

    Saves:
        'statistical_graph.png': A heatmap of the correlation matrix.
    """
    plt.figure(figsize=(12, 10))
    sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
    plt.title('Correlation Heatmap of the Dataset')
    plt.tight_layout()
    plt.savefig('statistical_graph.png')
    plt.show()

# Call the function
plot_statistical_graph(correlation_matrix)
```
